```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

Insert random data

```python
months = {
    "January": 31,
    "February": 28,
    "March": 31,
    "April": 30,
    "May": 31,
    "June": 30,
    "July": 31,
    "August": 31,
    "September": 30,
    "October": 31,
    "November": 30,
    "December": 31
}
month_data = {}

for month, days in months.items():
    month_data[f"{month}_data"] = list(np.random.randint(1, 10, size=days))

    # Convert the list to a NumPy array
    month_data[f"{month}_data"] = np.array(month_data[f"{month}_data"])
    print(f"{month}_data: {month_data[f'{month}_data']}")
```

##### Extension: Analyze internet usage ##### → peak days detection.
● Add-ons: Line plots, z-score outliers.

```python
for month in month_data:
  print(f"max_{month}_data :", np.max(month_data[month]))
  print(f"min_{month}_data :",np.min(month_data[month]))
  print(f"mean_{month}_data :", np.mean(month_data[month]))
  print(f"median_{month}_data :", np.median(month_data[month]))
  print(f"standard_deviation_{month}_data :", np.std(month_data[month]))
```

##### Graph work month-wise #####



line

```python
for month_name, days in month_data.items():
    days_in_month = len(days) # Get the actual number of days from the month_data
    plt.figure(figsize=(10, 4)) # Create a new figure for each month
    plt.plot(range(1, days_in_month + 1), days)
    plt.title(f"Daily WiFi Usage for {month_name.replace('_data', '')}")
    plt.xlabel("Day of Month")
    plt.ylabel("Usage")
    plt.grid(True)
    plt.show()
```

Z-score outlier

```python
for month, data in month_data.items():
    data = np.array(data)
    z_scores = (data - data.mean()) / data.std()
    days = np.arange(1, len(data) + 1)
    outliers = np.abs(z_scores) > 2

    plt.figure(figsize=(10, 4))
    plt.plot(days, z_scores, label="Z-score")
    plt.scatter(days[outliers], z_scores[outliers])
    plt.axhline(2)
    plt.axhline(-2)
    plt.title(f"Z-score Outliers for {month}")
    plt.xlabel("Day")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.show()


```

std deviation


```python
for month, data in month_data.items():
    data = np.array(data)
    mean, std = data.mean(), data.std()
    k = 2
    upper, lower = mean + k*std, mean - k*std
    days = np.arange(1, len(data) + 1)
    outliers = (data > upper) | (data < lower)

    plt.figure(figsize=(10, 4))
    plt.plot(days, data)
    plt.scatter(days[outliers], data[outliers])
    plt.axhline(mean, linestyle="--")
    plt.axhline(upper, linestyle=":")
    plt.axhline(lower, linestyle=":")
    plt.title(f"Std-Deviation Outliers ({month})")
    plt.xlabel("Day")
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()
```
